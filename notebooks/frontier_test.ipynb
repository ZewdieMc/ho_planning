Import

```python
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
```

Load Grid map

```python
image_map = "./test_map.png"

# Load grid map
image = Image.open(image_map).convert('L')
grid_map = np.array(image.getdata()).reshape(image.size[1],image.size[0]).astype('float64')
grid_map[grid_map == 205] = 0.5
grid_map[grid_map == 0] = 1
grid_map[grid_map > 205] = 0.0

# Show grid map
plt.matshow(grid_map)
plt.colorbar()
plt.show()
```
